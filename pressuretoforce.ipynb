```python
import numpy as np
import pandas as pd

ms = pd.read_csv('m_all.csv')
m = ms.to_numpy()
m = np.delete(m,0,1)

bs = pd.read_csv('b_all.csv')
b = bs.to_numpy()
b = np.delete(b,0,1)

#pandas gives an extra row that lists the sensor name, not normal python indexing
forces = np.zeros(7)
def ptoforces(pressures, xpos=3):
    forces = pressures*m[:,xpos-1]+b[:,xpos-1]
    return forces
```
